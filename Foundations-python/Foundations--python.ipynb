# Завдання 1. Основні типи даних

```python
# Прості змінні
a = "змінна з текстом"
b = 1         # int
b1 = 1.1      # float

# Список (list)
c = ["a", 1, 1.25, "Слово", a]

# Словник (dict)
d = {"a": "Слово", "b": 1, a: b}

# Кортеж (tuple)
e = ("a", a)

# Множина (set)
f = {"ss", a + str(b)}

# Вивід
print(a, b, b1)
print(c)
print(d)
print(e)
print(f)
```

# Завдання 2. Вбудовані константи та зарезервовані слова

```python
# Константи
print("Перша константа:", True)
print(f"Як можна так робити вивід? {False}")

# Перегляд зарезервованих слів
import keyword
print(keyword.kwlist[:10])  # Показати перші 10 зарезервованих слів
```

# Завдання 3. Вбудовані функції

```python
# Приклади вбудованих функцій
print(abs(-12.5), f"є рівним {abs(12.5)}", "і якщо порівняти то:", abs(-12.5) == abs(12.5))
print(len("Python"), f"довжина рядка 'Python'")
print(round(3.14159, 2), "округлення числа до 2 знаків")
```

# завдання 4. Цикли

```python
# Приклад циклу for
letters = ["a", "b", "c"]
for i in range(len(letters)):
    print(f"На позиції {i} знаходиться буква {letters[i]}")
else:
    print("Ця конструкція безглузда!")

# Приклад циклу while
count = 0
while count < 3:
    print(f"Цикл while, лічильник = {count}")
    count += 1
```

# Завдання 5. Розгалуження (if/else)

```python
from random import randint

A = randint(0, 1)
print(f"Значить A={A}" if A else f"Але може бути що A={A}")

B = 10
if B > 5:
    print("B більше 5")
elif B == 5:
    print("B дорівнює 5")
else:
    print("B менше 5")
```

# Завдання 6. Обробка помилок (try/except/finally)

```python
A = 0
try:
    print("Що буде якщо", 10/A, "?")
except Exception as e:
    print("Невже це помилка >", e)
finally:
    print("О це так на тобі!")
```

# Завдання 7. Контекст-менеджер (with)

```python
print("=== Робота з контекст-менеджером with ===")

try:
    with open("README.md", "r", encoding="utf-8") as file:
        for idx, line in enumerate(file):
            print(f"Рядок {idx + 1}: {line.strip()}")
            if idx >= 3:  # обмежуємо вивід першими 4 рядками
                print("...Далі ще рядки, але їх не показуємо...")
                break
except FileNotFoundError:
    print("Файл README.md не знайдено — створіть його у тій же папці, де скрипт.")
```

# Завдання 8. Lambda функції

```python
# Звичайна функція
def a_b_func(a, b):
    return a, b

# Lambda
this_is_lambda = lambda first, age: f'Цей код написав: {first}, Мені {age:10d} років'

print("Це просто функція:", a_b_func)
print("А це лямбда:", this_is_lambda)
print("Це її виклик:", this_is_lambda('Богдан', 1_00_000))
print(this_is_lambda(*a_b_func("a", 1)))
```
