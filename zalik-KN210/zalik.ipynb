```python
# Клас для пошуку мінімального значення
class MinValueFinder:
    def __init__(self, numbers):
        self.numbers = numbers
        self._min_value = None

    # Метод обробки даних
    def find_min(self):
        self._min_value = min(self.numbers)
        return self._min_value

    # Property для отримання результату
    @property
    def min_value(self):
        return self._min_value

    # Метод __str__ для виведення інформації про студента
    def __str__(self):
        return "Ім'я студента: Шульган Олексій\nГрупа: КН-210"


# Дані для створення об'єктів
data_lists = [
    [5, 3, 9, 1, 7],
    [12, -4, 6, 0, 8],
    [15, 22, 4, 10]
]

# Створення об'єктів у циклі
objects = []

for data in data_lists:
    obj = MinValueFinder(data)
    objects.append(obj)

# Виведення інформації
for obj in objects:
    print(obj)
    print(f"Мінімальне значення у списку {obj.numbers} дорівнює {obj.find_min()}")
    print(f"Значення через property: {obj.min_value}")
    print("-" * 45)
```
